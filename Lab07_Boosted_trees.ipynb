Perform GridSearchCV of the following three models on the provided training set (`X_train.csv` and `y_train.csv`)

1. Evaluate these models on the test set (`X_test.csv` and `y_test.csv`).

2. For each model, plot the feature importances

For `AdaBoostClassifier`, feature importances can be obtained by calling the `feature_importances` attribute after fitting the model.

For `XGBClassifier` and `LGBMClassifier`, feature importances can be obtained using the librarys `plot_importance` function. Here is a minimal example in XGBoost:

```python
import pandas as pd

X_train = pd.read_csv('X_train.csv', header=None)
X_test = pd.read_csv('X_test.csv', header=None)
y_train = pd.read_csv('y_train.csv', header=None)
y_test = pd.read_csv('y_test.csv', header=None)
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
from sklearn.ensemble import AdaBoostClassifier

ab = AdaBoostClassifier()

params_ab = {'n_estimators': [50, 100, 200],
             'learning_rate': [0.05, 0.1, 1.0]}

gridcv_ab = GridSearchCV(ab, params_ab, scoring='accuracy', cv=5)
gridcv_ab.fit(X_train, y_train)
```

```python
gridcv_ab.best_score_
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()

params_xgb = {'n_estimators': [50, 100, 200],
             'max_depth': [3, 5, 7],
             'learning_rate': [0.05, 0.1, 1.0]}

gridcv_xgb = GridSearchCV(xgb, params_xgb, scoring='accuracy', cv=5)
gridcv_xgb.fit(X_train, y_train)
```

```python
gridcv_xgb.best_score_
```

```python
from lightgbm import LGBMClassifier

lgbm = LGBMClassifier()

params_lgbm = {'n_estimators': [50, 100, 200],
               'max_depth': [3, 5, 7],
               'learning_rate': [0.05, 0.1, 1.0]}

gridcv_lgbm = GridSearchCV(lgbm, params_lgbm, scoring='accuracy', cv=5)
gridcv_lgbm.fit(X_train, y_train)
```

```python
gridcv_lgbm.best_score_
```

```python
best_model = gridcv_ab.best_estimator_
```

```python
from sklearn.metrics import classification_report

y_pred = best_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
feat_importances = pd.Series(gridcv_ab.best_estimator_.feature_importances_, index=X_train.columns)
features = feat_importances.sort_values(ascending=True)

ax = features.plot(kind='barh')
ax.bar_label(ax.containers[0], fmt='%.1f')

plt.title('Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Feature')

plt.show()
```
